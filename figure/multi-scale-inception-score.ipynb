```python

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
mpl.style.use('bmh')

mpl.rcParams['axes.linewidth'] = 2

ls = 'dotted'

fig, ax = plt.subplots(figsize=(5, 5))


x = [1, 2, 3]
y1 =  [3.53, 3.97, 4.15]
y2  = [2.95, 3.35, 3.70]
y1err = [0.03, 0.03, 0.05]
y2err = [0.02, 0.04, 0.04];
# standard error bars
ax.errorbar(x, y1, yerr=y1err, linestyle='-', marker='o', color='navy', markersize=8, label='HDGAN', linewidth=5)
ax.errorbar(x, y2, yerr=y2err, linestyle='--', marker='^', color='black', markersize=8, label='StackGAN', linewidth=5)

# tidy up the figure
ax.set_xlim((0.8, 3.2))
ax.set_ylim((2.6, 4.3))
#ax.set_title('Multi-scale inception score', fontsize=20)
#ax.grid(False)

plt.legend(loc=4, fontsize=20)
plt.xticks(x, ['64', '128', '256'], fontsize=15)
plt.yticks(fontsize=15)
ax.set_xlabel('Resolution', fontsize=20)
ax.set_ylabel('Inception score', fontsize=20)

ax.spines['bottom'].set_color('black')
ax.spines['top'].set_color('black')
ax.spines['left'].set_color('black')
ax.spines['right'].set_color('black')

plt.tight_layout()
#plt.show()
plt.savefig('multiscale_inception.pdf')
```
